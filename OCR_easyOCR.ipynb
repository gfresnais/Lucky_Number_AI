```python
# global variable for functions
kernel = np.ones((5,5),np.uint8)
```

Applies a closing which is a dilation followed by a erosion

```python
#closing - dilation followed by erosion
def closing(image):
    return cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
```

Applies an opening which is an erosion followed by a dilation

```python
#opening - erosion followed by dilation
def opening(image):
    return cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
```

Applies a top hat transform

```python
# Applies the top hat transform
def tophat(image):
    return cv2.morphologyEx(image, cv2.MORPH_TOPHAT, kernel)
```

Applies a black hat transform

```python
# Applies the black hat transform
def blackhat(image):
    return cv2.morphologyEx(image, cv2.MORPH_BLACKHAT, kernel)
```

Applies a gradient

```python
# Applies a gradient
def gradient(image):
    return cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)
```

## Main program

Defining our image variables

```python
# images vars
img_dir = "App/res/images_AI/";
img_names = {'yellow.png', 'red.png', 'green.png', 'purple.png'}
```

```python
# ranges for color detection
# detects darker colors
lower_range = np.array([0,0,0])
upper_range = np.array([180,170,150])
```

Starting a loop from 1 to 20 which corresponds to each number available in the game

Adding a second loop inside to browse the different tokens' colors

Reads an image thanks to the vars

```python
# Will count the number of correct guesses
cpt_red = 0
cpt_purple = 0
cpt_green = 0
cpt_yellow = 0

reader = easyocr.Reader(['en'])

for i in range(1, 21):
    for name in img_names: 
        # opens the image file
        image = cv2.imread(img_dir + str(i) + '/' + str(i) + name)
        
        # get the image dimensions
        h, w, c = image.shape
        
        # print the dimensions (width*height)
        print("Image dimensions: " + str(w) + "x" + str(h))
        
        # removes the noise (closing)
        blurred = closing(image)
        
        # convert to hsv color space
        hsv = cv2.cvtColor(blurred, cv2.COLOR_BGR2HSV)
        
        # apply a mask to the hsv image
        mask = cv2.inRange(hsv, lower_range, upper_range)
        
        # apply the mask to the original image
        bitwise = cv2.bitwise_and(image, image, mask=mask)
        
        # convert the image back to bgr color space
        img_to_bgr = cv2.cvtColor(bitwise, cv2.COLOR_HSV2BGR)
        
        # copy the image so it will be a cleaned image for tesseract
        img = img_to_bgr.copy()
        
        # split the copied image into boxes
        boxes = reader.detect(img)
        
        # draw each box
        for b in boxes:
            if(len(b) > 0):
                b = b[0]
                img = cv2.rectangle(img, (int(b[0]), int(b[2])), (int(b[1]), int(b[3])), (255, 255, 255), 2)
        
        # gets the result from easyocr
        result = reader.readtext(img_to_bgr, allowlist="0123456789")
        
        # will print the boxes, the prediction and the confidence
        print(result)
        
        # draw the chart containing the image with boxes
        plt.imshow(img)
        plt.title(name)
        plt.show()
        
        # initialize res to 0 as a failsafe
        res = "0"
        
        # if a result was found by easyocr (length > 0), res will be equal to the predicted result
        if len(result) > 0:
            if result[0][1].isdigit():
                res = result[0][1]
        
        # removes the useless whitespaces
        res = res.strip()
        res = res.replace(" ","")
        
        print(name + '(' + str(i) + '):' + res)
        
        if( int(res) == i ):
            if( name == 'yellow.png' ): 
                cpt_purple += 1
            if( name == 'yellow.png' ): 
                cpt_yellow += 1
            if( name == 'green.png' ): 
                cpt_green += 1
            if( name == 'red.png' ): 
                cpt_red += 1
        print("-----------------------------")
```

```python
print("Number of correct red = " + str(cpt_red) + " / 20")
print("Number of correct purple = " + str(cpt_purple) + " / 20")
print("Number of correct green = " + str(cpt_green) + " / 20")
print("Number of correct yellow = " + str(cpt_yellow) + " / 20")
```
